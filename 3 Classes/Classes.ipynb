### Creating a user defined class



```python
class Person:

    # Always define constructor function to define attributes
    def __init__(self, name, age, gender, occupation):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation

    # Add a user defined method
    def introduce(self):
        print(f'Hi there my name is {self.name} .')
        print(f'I am {self.age} years old .')
        print(f'I am a {self.gender} .')
        print(f'I work as a {self.occupation} .')
```

### Creating the objects for class Person

```python
p1 = Person("Sarthak", 23, "Male", "Engineer")
```

### Bank Account example



```python
class Account:

    def __init__(self, ac_no, name, bal=0):
        self.ac_no = ac_no
        self.name = name 
        self.bal = bal

    def current_balance(self):
        print(f"Current Balance is : {self.bal} INR")

    def deposit(self, amt):
        # Money will get added in the user account
        self.bal = self.bal + amt
        # Print the balance 
        print(f"Amount of {amt} INR got deposited in account : {self.ac_no}")
        self.current_balance()

    def withdraw(self, amt):
        if amt <= self.bal:
            self.bal = self.bal - amt 
            print(f"Amount of {amt} INR got withdrawn account : {self.ac_no}")
        else:
            print("Insufficient balance for Withdrawal")
        self.current_balance()

    def transfer(self, ac2, amt):
        if amt <= self.bal:
            self.bal = self.bal - amt
            ac2.bal = ac2.bal + amt 
            print(f"Amount of {amt} INR got transferred from {self.ac_no} to {ac2.ac_no}")
        else:
            print("Insufficient Balance for Transfer")
        self.current_balance()
```
